```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path='SampleSuperstore.csv'
df=pd.read_csv(path)
print("Dataset Loaded")
```

```python
#Checking for duplicate values
df.duplicated().sum()
```

```python
#Droping the duplicate values
df.drop_duplicates(inplace=True)
```

```python
#Relationship analysis and visualizing the dataset
correlation=df.corr()
```

```python
#Heatmap
sns.heatmap(correlation,annot=True)
```

```python
#Pairplot
sns.pairplot(df)
```

```python
#Boxplot
sns.boxplot(df['Profit'])
```

```python
#Relation plot
sns.relplot(x='Profit',y='Discount',data=df)
```

```python
vis=df.groupby(['Ship Mode'],as_index=False)['Profit'].sum()
sns.barplot(vis['Ship Mode'],vis['Profit'])
```

```python
disc=df.groupby(['Ship Mode'],as_index=False)['Discount'].sum()
sns.barplot(disc['Ship Mode'],disc['Discount'])
```

```python
sns.distplot(df['Profit'])
```

```python
sns.pairplot(df, hue="Ship Mode")
```

```python
data = df.groupby(['Segment'],as_index = False).sum()
data
```

```python
sns.barplot(x = 'Segment', y = 'Quantity', data = data)
```

```python
#Calculating region wise profit
df.groupby(['Region'], as_index = False, sort = True)['Profit'].mean()
```

```python
#Pie Chart
region=['Central','East','South','West']
profit=[17.093,32.135,28.858,33.849]
plt.pie(profit,labels=region,autopct='%1.2f%%')
```

```python
sns.lineplot(y='Profit',x='Discount',data=df)
```


From above we can conclude that if discount is increased the profit is decreased

Conclusion: 1.Evaluating the different regions the East region has the highest profit while Central region has lowest Profit. 2.The Consumer segment purchases more quantity while home office segment purchases least quantity. 3.The Standard Class is Earning the more Profit as it is offering more Discounts and having huge sales then other classes.
